```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

# Define the logistic function
def logistic(x, L, k, x0):
    return L / (1 + np.exp(-k * (x - x0)))

# Your experimental data (replace with your actual values)
x_data = np.array([0, 1, 2, 3, 4, 5, 10])
y_data = np.array([0.1, 0.2, 0.4, 0.65, 0.85, 0.95, 0.99])

# Fit the logistic function to your data
popt, pcov = curve_fit(logistic, x_data, y_data, p0=[1, 1, 1])
L_fit, k_fit, x0_fit = popt

# Generate smooth curve for plotting the fit
x_fit = np.linspace(min(x_data), max(x_data), 500)
y_fit = logistic(x_fit, *popt)

# Plot experimental data and fitted curve
plt.scatter(x_data, y_data, label='Experimental data', color='red')
plt.plot(x_fit, y_fit, label=f'Logistic fit\nL={L_fit:.2f}, k={k_fit:.2f}, xâ‚€={x0_fit:.2f}')
plt.title('Logistic Fit to PFAS breakthrough curve')
plt.xlabel('time')
plt.ylabel('C/Co')
plt.legend()
plt.grid(True)
plt.show()
```
